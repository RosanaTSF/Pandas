```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
from IPython.display import Image, display, clear_output

# === 1. Carregando os dados ===
csv_path = '/kaggle/input/cat-breeds/dataset_stats.csv'
df = pd.read_csv(csv_path)
clear_output(wait=True)

# === 2. Mostra as primeiras linhas ===
print("Primeiras linhas da tabela:")
display(df.head())

print("\nVerificando valores nulos:")
print(df.isnull().sum())

# === 3. Exibe as ra√ßas com mais imagens ===
print("\nTop 10 ra√ßas com mais imagens:")
top_breeds = df.sort_values(by='image_count', ascending=False).head(10)
display(top_breeds)

# === 4. Mostra as ra√ßas com menos imagens ===
print("\nTop 10 ra√ßas com MENOS imagens:")
bottom_breeds = df.sort_values(by='image_count', ascending=True).head(10)
display(bottom_breeds)

# === 5. Gr√°fico da distribui√ß√£o das quantidades de imagens ===
plt.figure(figsize=(10, 5))
plt.hist(df['image_count'], bins=20, color='purple', edgecolor='black')
plt.title('Distribui√ß√£o de Quantidade de Imagens por Ra√ßa')
plt.xlabel('Quantidade de imagens')
plt.ylabel('N√∫mero de ra√ßas')
plt.grid(True)
plt.show()

# === 6. Gr√°fico das 10 ra√ßas com menos imagens ===
plt.figure(figsize=(10, 5))
plt.bar(bottom_breeds['class'], bottom_breeds['image_count'], color='orange')
plt.title('Top 10 Ra√ßas com Menos Imagens')
plt.xlabel('Ra√ßa')
plt.ylabel('Quantidade de Imagens')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# === 7. Mostra uma imagem por ra√ßa ===
print("\nMostrando uma imagem por ra√ßa:")
base_path = '/kaggle/input/cat-breeds/cat-breeds/cat-breeds'
breeds = os.listdir(base_path)

for breed in breeds:
    breed_path = os.path.join(base_path, breed)
    if os.path.isdir(breed_path):
        for image_filename in os.listdir(breed_path):
            if image_filename.endswith(('.jpg', '.png')):
                image_path = os.path.join(breed_path, image_filename)
                print(f"Ra√ßa: {breed}")
                display(Image(filename=image_path))
                break  # Mostra s√≥ a primeira imagem


# Filtra as ra√ßas que t√™m menos de 10 imagens
poucas_imagens = df[df['image_count'] < 10]

# Mostra o n√∫mero de ra√ßas com poucas imagens
print(f"Quantidade de ra√ßas com menos de 10 imagens: {len(poucas_imagens)}")

# Mostra as ra√ßas e suas quantidades
print("\nRa√ßas com menos de 10 imagens:")
display(poucas_imagens)
```
