```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
from IPython.display import Image, display, clear_output

# === 1. Carregando os dados ===
csv_path = '/kaggle/input/cat-breeds/dataset_stats.csv'
df = pd.read_csv(csv_path)
clear_output(wait=True)

# === 2. Mostra as primeiras linhas ===
print("Primeiras linhas da tabela:")
display(df.head())

print("\nVerificando valores nulos:")
print(df.isnull().sum())

# === 3. Exibe as raças com mais imagens ===
print("\nTop 10 raças com mais imagens:")
top_breeds = df.sort_values(by='image_count', ascending=False).head(10)
display(top_breeds)

# === 4. Mostra as raças com menos imagens ===
print("\nTop 10 raças com MENOS imagens:")
bottom_breeds = df.sort_values(by='image_count', ascending=True).head(10)
display(bottom_breeds)

# === 5. Gráfico da distribuição das quantidades de imagens ===
plt.figure(figsize=(10, 5))
plt.hist(df['image_count'], bins=20, color='purple', edgecolor='black')
plt.title('Distribuição de Quantidade de Imagens por Raça')
plt.xlabel('Quantidade de imagens')
plt.ylabel('Número de raças')
plt.grid(True)
plt.show()

# === 6. Gráfico das 10 raças com menos imagens ===
plt.figure(figsize=(10, 5))
plt.bar(bottom_breeds['class'], bottom_breeds['image_count'], color='orange')
plt.title('Top 10 Raças com Menos Imagens')
plt.xlabel('Raça')
plt.ylabel('Quantidade de Imagens')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# === 7. Mostra uma imagem por raça ===
print("\nMostrando uma imagem por raça:")
base_path = '/kaggle/input/cat-breeds/cat-breeds/cat-breeds'
breeds = os.listdir(base_path)

for breed in breeds:
    breed_path = os.path.join(base_path, breed)
    if os.path.isdir(breed_path):
        for image_filename in os.listdir(breed_path):
            if image_filename.endswith(('.jpg', '.png')):
                image_path = os.path.join(breed_path, image_filename)
                print(f"Raça: {breed}")
                display(Image(filename=image_path))
                break  # Mostra só a primeira imagem


# Filtra as raças que têm menos de 10 imagens
poucas_imagens = df[df['image_count'] < 10]

# Mostra o número de raças com poucas imagens
print(f"Quantidade de raças com menos de 10 imagens: {len(poucas_imagens)}")

# Mostra as raças e suas quantidades
print("\nRaças com menos de 10 imagens:")
display(poucas_imagens)
```
